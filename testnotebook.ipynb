```python
import requests
import pandas as pd
import io
import json
print("Import erfolgreich")
```

```python
r = requests.get('https://api.acleddata.com/acled/read?terms=accept').json()
print(r)
```

```python
content = json.dumps(r,indent=4,sort_keys=True)
#print(content)
```

```python
accleddf = pd.read_json(content)
accleddf = accleddf.data
print(accleddf)
```
